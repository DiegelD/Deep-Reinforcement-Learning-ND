```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
%matplotlib inline

import torch
import torch.nn.functional as F
import torch.optim as optim
```

## Model Comparison

```python
# Open the file in read mode
objectRep = open("./saved_models/score_Critic256_256_128_Actor256_Buffer1000000_Batch128_Noise_1e6.pkl", "rb")
print("Unpickling...")
# Unpickle the objects
score1 = pickle.load(objectRep)
for i in score1:score1 = i
print("Loaded...")
# Open the file in read mode
objectRep = open("./saved_models/score_Actor400_300_Critic256_256_128_Buffer1000000_Batch128_Noise_1e6.pkl", "rb")
print("Unpickling...")
# Unpickle the objects
score2 = pickle.load(objectRep)
for i in score2:score2 = i
print("Loaded...")
# Open the file in read mode
objectRep = open("./saved_models/score_Actor400_300_Critic200_200_Buffer1000000_Batch128_Noise_1e6.pkl", "rb")
print("Unpickling...")
# Unpickle the objects
score3  = pickle.load(objectRep)
for i in score3:score3 = i
print("Loaded...")

# plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(score1)), score1)
plt.plot(np.arange(len(score2)), score2)
plt.plot(np.arange(len(score3)), score3)
# Solved Line
reached_line = []
for i in range(550): reached_line.append(30)
plt.plot(np.arange(550), reached_line, 'k--')
ax.text(30, 31, r'Environment Solved', fontsize=11)


saving_name = 'ModelComparison'

plt.ylabel('Score')
plt.xlabel('Episode')
plt.title('Model Comparison')
ax.legend(['1 ', '2','3'])
plt.show()
fig.savefig('./img/Scores_'+str(saving_name)+'.jpg', dpi=400)
```
